```python
!pip install torch torchvision numpy scikit-learn transformers matplotlib > /dev/null 2>&1

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import numpy as np
import random
import matplotlib.pyplot as plt
from collections import deque

# --- CONFIGURATION ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.manual_seed(42)
np.random.seed(42)

# --- MODELS ---
class TrafficCNN(nn.Module):
    def __init__(self):
        super(TrafficCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, 3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1)
        self.fc1 = nn.Linear(32 * 16 * 16, 128)
        self.fc2 = nn.Linear(128, 1)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 32 * 16 * 16)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return torch.sigmoid(x)

class TrafficLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(TrafficLSTM, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(device)
        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out

class TrafficVAE(nn.Module):
    def __init__(self, input_dim, latent_dim):
        super(TrafficVAE, self).__init__()
        self.fc1 = nn.Linear(input_dim, 128)
        self.fc21 = nn.Linear(128, latent_dim)
        self.fc22 = nn.Linear(128, latent_dim)
        self.fc3 = nn.Linear(latent_dim, 128)
        self.fc4 = nn.Linear(128, input_dim)

    def encode(self, x):
        h1 = F.relu(self.fc1(x))
        return self.fc21(h1), self.fc22(h1)

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z):
        h3 = F.relu(self.fc3(z))
        return torch.sigmoid(self.fc4(h3))

    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        return self.decode(z), mu, logvar

class SocialNLP(nn.Module):
    def __init__(self, vocab_size, embed_dim, num_heads, hidden_dim):
        super(SocialNLP, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.transformer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=num_heads, dim_feedforward=hidden_dim)
        self.fc = nn.Linear(embed_dim, 1)

    def forward(self, x):
        x = self.embedding(x)
        x = x.permute(1, 0, 2)
        x = self.transformer(x)
        x = x.mean(dim=0)
        return torch.sigmoid(self.fc(x))

class RLAgent:
    def __init__(self, state_size, action_size):
        self.state_size = state_size
        self.action_size = action_size
        self.memory = deque(maxlen=2000)
        self.gamma = 0.95
        self.epsilon = 1.0
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.model = self._build_model()
        self.criterion = nn.MSELoss()
        self.optimizer = optim.Adam(self.model.parameters(), lr=0.001)

    def _build_model(self):
        return nn.Sequential(
            nn.Linear(self.state_size, 24), nn.ReLU(),
            nn.Linear(24, 24), nn.ReLU(),
            nn.Linear(24, self.action_size)
        ).to(device)

    def act(self, state):
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        state = torch.FloatTensor(state).to(device)
        act_values = self.model(state)
        return torch.argmax(act_values).item()

    def replay(self, batch_size):
        if len(self.memory) < batch_size: return
        minibatch = random.sample(self.memory, batch_size)
        for state, action, reward, next_state, done in minibatch:
            target = reward
            if not done:
                next_state_t = torch.FloatTensor(next_state).to(device)
                target = reward + self.gamma * torch.max(self.model(next_state_t)).item()
            state_t = torch.FloatTensor(state).to(device)
            target_f = self.model(state_t).detach().clone()
            target_f[action] = target
            self.optimizer.zero_grad()
            output = self.model(state_t)
            loss = self.criterion(output, target_f)
            loss.backward()
            self.optimizer.step()
        if self.epsilon > self.epsilon_min: self.epsilon *= self.epsilon_decay

# --- SIMULATION ---
class MockTrafficEnv:
    def __init__(self):
        self.state_size = 5
        self.action_size = 4
        self.time_step = 0

    def reset(self):
        self.time_step = 0
        return np.random.rand(self.state_size)

    def step(self, action):
        self.time_step += 1
        next_state = np.random.rand(self.state_size)
        # Simulation Logic: Trade-offs
        congestion = next_state[0]
        efficiency = 1.0 - congestion
        # If action is "Green Light" (even), higher emissions but better flow
        emissions = congestion * 0.8 + (0.2 if action % 2 == 0 else 0.05)
        fairness = 1.0 - abs(next_state[1] - next_state[2])

        # Reward Function balancing objectives
        reward = (0.5 * efficiency) - (0.3 * emissions) + (0.2 * fairness)
        done = self.time_step >= 100
        return next_state, reward, done, {"eff": efficiency, "emit": emissions, "fair": fairness}

    def get_data(self):
        return (torch.randn(1, 3, 64, 64).to(device),
                torch.randn(1, 10, 5).to(device),
                torch.randint(0, 1000, (1, 20)).to(device))

# --- MAIN LOOP WITH PLOTTING ---
def run_project():
    print("Initializing AI Traffic System...")
    # Instantiate
    cnn = TrafficCNN().to(device)
    lstm = TrafficLSTM(5, 32, 2, 1).to(device)
    vae = TrafficVAE(5, 2).to(device)
    nlp = SocialNLP(1000, 32, 2, 64).to(device)
    agent = RLAgent(5, 4)
    env = MockTrafficEnv()

    # Metrics Storage
    history = {'reward': [], 'efficiency': [], 'emissions': [], 'fairness': []}
    state = env.reset()

    print("Training Pipeline (Simulating 100 Steps)...")
    for t in range(100):
        # 1. Multi-modal Input Processing
        img, sensor_seq, text = env.get_data()
        _ = cnn(img) # Vision processing
        pred_congestion = lstm(sensor_seq).item() # Time-series

        # 2. Anomaly Detection
        snapshot = torch.FloatTensor(state).to(device).unsqueeze(0)
        recon, _, _ = vae(snapshot)
        anomaly_loss = F.mse_loss(recon, snapshot).item()

        # 3. NLP Event Detection
        event_prob = nlp(text).item()

        # 4. State Update & RL Action
        state[0] = (state[0] + pred_congestion) / 2.0 # Fuse data
        action = agent.act(state)

        # 5. Environment Step
        next_state, reward, done, info = env.step(action)
        agent.memory.append((state, action, reward, next_state, done))
        agent.replay(16)

        # 6. Log Metrics
        history['reward'].append(reward)
        history['efficiency'].append(info['eff'])
        history['emissions'].append(info['emit'])
        history['fairness'].append(info['fair'])
        state = next_state

    # --- PLOTTING RESULTS ---
    print("Generating Report Plots...")
    plt.figure(figsize=(15, 5))

    # Plot 1: Learning Curve (Reward)
    plt.subplot(1, 3, 1)
    plt.plot(history['reward'], color='blue', label='RL Reward')
    plt.title('AI Training Convergence')
    plt.xlabel('Simulation Step')
    plt.ylabel('Total Reward')
    plt.legend()
    plt.grid(True)

    # Plot 2: Conflicting Requirements (Efficiency vs Emissions)
    plt.subplot(1, 3, 2)
    plt.plot(history['efficiency'], color='green', label='Efficiency (Flow)')
    plt.plot(history['emissions'], color='red', linestyle='--', label='Emissions (CO2)')
    plt.title('Trade-off: Efficiency vs Sustainability')
    plt.xlabel('Simulation Step')
    plt.legend()
    plt.grid(True)

    # Plot 3: Fairness Metric
    plt.subplot(1, 3, 3)
    plt.plot(history['fairness'], color='purple', label='Gini Fairness')
    plt.title('Fairness Score (Avoiding Bias)')
    plt.xlabel('Simulation Step')
    plt.ylim(0, 1.1)
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig('traffic_optimization_results.png')
    plt.show()

    print("\nSUCCESS: Pipeline execution complete.")
    print("1. 'traffic_optimization_results.png' saved.")
    print(f"2. Final Efficiency: {np.mean(history['efficiency']):.2f}")
    print(f"3. Final Fairness: {np.mean(history['fairness']):.2f}")

if __name__ == "__main__":
    run_project()
```
