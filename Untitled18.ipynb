```python
# ========= 2.  SYNTHETIC DATA =========
def gini(x):                       # local replacement for scipy.stats.gini
    x = np.array(x, dtype=float)
    n = len(x)
    index = np.arange(1, n + 1)
    return (2 * np.sum(index * np.sort(x))) / (n * np.sum(x)) - (n + 1) / n

def generate_synthetic_data(n_days=14, freq="5min"):
    dt = pd.date_range("2025-01-01", periods=n_days*24*12, freq=freq)
    n = len(dt)
    base = 50 + 30*np.sin(np.linspace(0, n_days*2*np.pi, n))
    vol  = np.clip(base + np.random.normal(0, 5, n), 0, 200).astype(int)
    speed = np.clip(60 - 0.4*vol + np.random.normal(0, 3, n), 5, 80)
    co2   = 120 + 0.5*(80-speed).clip(0, 50) + np.random.normal(0, 5, n)
    accident = np.random.binomial(1, 0.005, n)
    emg      = np.random.binomial(1, 0.01, n)
    weather  = np.random.choice(["Clear","Rain","Fog"], n, p=[0.7,0.2,0.1])
    tweets   = np.random.choice([-1,0,1], n, p=[0.1,0.7,0.2])
    df = pd.DataFrame({
        "datetime": dt,
        "volume": vol,
        "speed": speed,
        "co2": co2,
        "weather": weather,
        "accident": accident,
        "emergency": emg,
        "tweet_sentiment": tweets
    })
    return df

df = generate_synthetic_data()
df.to_csv(os.path.join(CFG["data_dir"], "synthetic_traffic.csv"), index=False)
print(df.head())
```

```python
# ========= 3.  PRE-PROCESS =========
def make_sequences(df, seq_len=24, target="volume"):
    scaler = MinMaxScaler()
    vals = scaler.fit_transform(df[[target]].values).flatten()
    X, y = [], []
    for i in range(len(vals)-seq_len):
        X.append(vals[i:i+seq_len])
        y.append(vals[i+seq_len])
    return np.array(X), np.array(y), scaler

X, y, scaler = make_sequences(df, CFG["seq_len"], target="volume")
split = int(0.8*len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

train_ds = torch.utils.data.TensorDataset(torch.Tensor(X_train).unsqueeze(-1),
                                          torch.Tensor(y_train))
test_ds  = torch.utils.data.TensorDataset(torch.Tensor(X_test).unsqueeze(-1),
                                          torch.Tensor(y_test))
train_loader = DataLoader(train_ds, batch_size=CFG["batch_size"], shuffle=True)
test_loader  = DataLoader(test_ds,  batch_size=CFG["batch_size"])
print("Train batches:", len(train_loader), "| Test batches:", len(test_loader))
```

```python
# ========= 4.  VAE =========
class VAE(nn.Module):
    def __init__(self, in_dim, latent):
        super().__init__()
        self.enc = nn.Sequential(nn.Linear(in_dim, 64), nn.ReLU(),
                                 nn.Linear(64, 32), nn.ReLU())
        self.mu  = nn.Linear(32, latent)
        self.logv= nn.Linear(32, latent)
        self.dec = nn.Sequential(nn.Linear(latent, 32), nn.ReLU(),
                                 nn.Linear(32, 64), nn.ReLU(),
                                 nn.Linear(64, in_dim))
    def encode(self, x):
        h = self.enc(x)
        return self.mu(h), self.logv(h)
    def reparam(self, mu, logv):
        std = torch.exp(0.5*logv)
        eps = torch.randn_like(std)
        return mu + eps*std
    def decode(self, z):
        return self.dec(z)
    def forward(self, x):
        mu, logv = self.encode(x)
        z = self.reparam(mu, logv)
        return self.decode(z), mu, logv

def vae_loss(recon, x, mu, logv):
    recon_loss = nn.MSELoss()(recon, x)
    kl = -0.5 * torch.sum(1 + logv - mu.pow(2) - logv.exp(), dim=1).mean()
    return recon_loss + kl

vae_model = VAE(CFG["seq_len"], CFG["latent_dim"]).to(device)
opt_vae = torch.optim.Adam(vae_model.parameters(), lr=CFG["lr"])

for epoch in range(CFG["epochs_vae"]):
    vae_model.train()
    tot=0
    for xb,_ in train_loader:
        xb = xb.squeeze(-1).to(device)
        recon, mu, logv = vae_model(xb)
        loss = vae_loss(recon, xb, mu, logv)
        opt_vae.zero_grad(); loss.backward(); opt_vae.step()
        tot+=loss.item()
    print(f"VAE epoch {epoch+1} loss={tot/len(train_loader):.4f}")
```

```python
# ========= 5.  LSTM =========
class LSTMNet(nn.Module):
    def __init__(self, input_size=1, hidden=64, layers=2):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden, layers, batch_first=True)
        self.fc   = nn.Linear(hidden, 1)
    def forward(self, x):
        out, _ = self.lstm(x)
        return self.fc(out[:, -1, :])

lstm_model = LSTMNet().to(device)
opt_lstm = torch.optim.Adam(lstm_model.parameters(), lr=CFG["lr"])
crit_lstm = nn.MSELoss()

for epoch in range(CFG["epochs_lstm"]):
    lstm_model.train()
    tot=0
    for xb,yb in train_loader:
        xb,yb = xb.to(device), yb.to(device)
        pred = lstm_model(xb).squeeze()
        loss = crit_lstm(pred, yb)
        opt_lstm.zero_grad(); loss.backward(); opt_lstm.step()
        tot+=loss.item()
    print(f"LSTM epoch {epoch+1} MSE={tot/len(train_loader):.4f}")
```

```python
# ========= 6.  RL ENV =========
class TrafficLightEnv:
    def __init__(self, df, lstm_model, scaler):
        self.df=df.reset_index(drop=True); self.lstm=lstm_model; self.scaler=scaler
        self.idx=CFG["seq_len"]; self.done=False
    def reset(self):
        self.idx=CFG["seq_len"]; self.done=False; return self._state()
    def _state(self):
        row=self.df.iloc[self.idx]
        vol=row["volume"]
        seq=self.scaler.transform([[vol]])[0][0]
        with torch.no_grad():
            x=torch.Tensor([[seq]*CFG["seq_len"]]).unsqueeze(-1).to(device)
            pred=self.lstm(x).cpu().item()
        return np.array([vol, pred, row["co2"], row["emergency"]], dtype=np.float32)
    def step(self, action):
        row=self.df.iloc[self.idx]
        delay_base=row["volume"]*0.5
        if action==0: delay=delay_base
        elif action==1: delay=delay_base*0.95
        else: delay=delay_base*1.05
        reward= -(CFG["lambda_time"]*delay + CFG["lambda_co2"]*row["co2"])
        if row["emergency"]: reward+=5
        self.idx+=1; self.done=self.idx>=len(self.df)-1
        return self._state(), reward, self.done, {}

env=TrafficLightEnv(df, lstm_model, scaler)
```
