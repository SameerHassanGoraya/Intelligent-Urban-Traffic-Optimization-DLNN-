```python
!pip install -q transformers

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
import torchvision.utils as vutils
import numpy as np
import matplotlib.pyplot as plt
from torchvision.datasets import Flowers102
from torch.utils.data import DataLoader
from transformers import DistilBertTokenizer, DistilBertModel

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

BATCH_SIZE = 64
IMAGE_SIZE = 64
Z_DIM = 100
EMBED_DIM = 768
LR_G = 0.0001
LR_D = 0.0004
EPOCHS = 5

transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

print("Loading Text Model...")
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
text_encoder = DistilBertModel.from_pretrained('distilbert-base-uncased').to(device)
text_encoder.eval()

class RealFlowerDataset(torch.utils.data.Dataset):
    def __init__(self, root, split='test', transform=None):
        self.dataset = Flowers102(root=root, split=split, download=True, transform=transform)

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        img, label = self.dataset[idx]
        txt = f"flower class {label}"
        tokens = tokenizer(txt, return_tensors="pt", padding='max_length', truncation=True, max_length=10)
        return img, tokens['input_ids'].squeeze(0), tokens['attention_mask'].squeeze(0)

print("Downloading Images...")
dataset = RealFlowerDataset(root='./data', split='test', transform=transform)
dataloader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)

def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)

class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.ngf = 64
        self.fc = nn.Sequential(
            nn.Linear(Z_DIM + EMBED_DIM, 4 * 4 * self.ngf * 8),
            nn.BatchNorm1d(4 * 4 * self.ngf * 8),
            nn.ReLU(True)
        )
        self.conv = nn.Sequential(
            nn.ConvTranspose2d(self.ngf * 8, self.ngf * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(self.ngf * 4),
            nn.ReLU(True),
            nn.ConvTranspose2d(self.ngf * 4, self.ngf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(self.ngf * 2),
            nn.ReLU(True),
            nn.ConvTranspose2d(self.ngf * 2, self.ngf, 4, 2, 1, bias=False),
            nn.BatchNorm2d(self.ngf),
            nn.ReLU(True),
            nn.ConvTranspose2d(self.ngf, 3, 4, 2, 1, bias=False),
            nn.Tanh()
        )

    def forward(self, noise, text_embed):
        x = torch.cat([noise, text_embed], 1)
        x = self.fc(x)
        x = x.view(-1, self.ngf * 8, 4, 4)
        return self.conv(x)

class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.ndf = 64
        self.img_conv = nn.Sequential(
            nn.Conv2d(3, self.ndf, 4, 2, 1, bias=False),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(self.ndf, self.ndf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(self.ndf * 2),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(self.ndf * 2, self.ndf * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(self.ndf * 4),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(self.ndf * 4, self.ndf * 8, 4, 2, 1, bias=False),
            nn.BatchNorm2d(self.ndf * 8),
            nn.LeakyReLU(0.2, inplace=True)
        )
        self.txt_fc = nn.Sequential(
            nn.Linear(EMBED_DIM, 128),
            nn.LeakyReLU(0.2, inplace=True)
        )
        self.joint_conv = nn.Sequential(
            nn.Conv2d(self.ndf * 8 + 128, self.ndf * 8, 1, 1, 0, bias=False),
            nn.BatchNorm2d(self.ndf * 8),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(self.ndf * 8, 1, 4, 1, 0, bias=False),
            nn.Sigmoid()
        )

    def forward(self, img, text_embed):
        img_out = self.img_conv(img)
        txt_out = self.txt_fc(text_embed)
        txt_out = txt_out.view(-1, 128, 1, 1).repeat(1, 1, 4, 4)
        return self.joint_conv(torch.cat([img_out, txt_out], 1)).view(-1, 1).squeeze(1)

netG = Generator().to(device)
netD = Discriminator().to(device)
netG.apply(weights_init)
netD.apply(weights_init)

criterion = nn.BCELoss()
optD = optim.Adam(netD.parameters(), lr=LR_D, betas=(0.5, 0.999))
optG = optim.Adam(netG.parameters(), lr=LR_G, betas=(0.5, 0.999))

print("Starting Training...")
loss_G, loss_D = [], []

for epoch in range(EPOCHS):
    for i, data in enumerate(dataloader):
        imgs, ids, mask = data
        imgs = imgs.to(device)
        b_size = imgs.size(0)

        with torch.no_grad():
            txt_embed = text_encoder(ids.to(device), mask.to(device)).last_hidden_state[:, 0, :]

        netD.zero_grad()
        label = torch.full((b_size,), 1.0, device=device)
        errD_real = criterion(netD(imgs, txt_embed), label)
        errD_real.backward()

        noise = torch.randn(b_size, Z_DIM, device=device)
        fake = netG(noise, txt_embed)
        label.fill_(0.0)
        errD_fake = criterion(netD(fake.detach(), txt_embed), label)
        errD_fake.backward()
        optD.step()

        netG.zero_grad()
        label.fill_(1.0)
        errG = criterion(netD(fake, txt_embed), label)
        errG.backward()
        optG.step()

    print(f'Epoch {epoch+1}/{EPOCHS} Loss_D: {errD_real+errD_fake:.4f} Loss_G: {errG.item():.4f}')
    loss_G.append(errG.item())
    loss_D.append(errD_real.item() + errD_fake.item())

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(loss_G, label="G")
plt.plot(loss_D, label="D")
plt.legend()
plt.subplot(1,2,2)
plt.axis("off")
plt.imshow(np.transpose(vutils.make_grid(fake.detach().cpu()[:16], padding=2, normalize=True), (1,2,0)))
plt.show()
```
